```python
# Generate data from GMM

N = 200
mu1, sigma1, pi1 = -2.5, 1, 0.4
mu2, sigma2, pi2 = 1.5, 1, 0.6

indices = np.random.choice([0, 1], size=N, p=[pi1, pi2])

# Générer des points à partir des composantes
points = np.concatenate([
    np.random.normal(mu1, sigma1, size=np.sum(indices == 0)),
    np.random.normal(mu2, sigma2, size=np.sum(indices == 1))
])

# Plot histograms with varying number of bins
Nbs = [10, 15, 20, 50, 100, 200]
bandwidths = [(max(points)-min(points))/Nb for Nb in Nbs]
num_subplots = len(Nbs)
fig, axes = plt.subplots(num_subplots, 1, figsize=(5, 3*num_subplots), sharex=True)
x = np.linspace(min(points), max(points), 1000)

for i, h in enumerate(bandwidths):

   
    pdf1 = pi1 * (1/(sigma1 * np.sqrt(2 * np.pi))) * np.exp(-(x - mu1)**2 / (2 * sigma1**2))
    pdf2 = pi2 * (1/(sigma2 * np.sqrt(2 * np.pi))) * np.exp(-(x - mu2)**2 / (2 * sigma2**2))
    pdf3 = pdf1 + pdf2
    axes[i].plot(x, pdf1, 'g--')
    axes[i].plot(x, pdf2, 'b--')
    axes[i].plot(x, pdf3, 'r')

    hist, bins = np.histogram(points, bins=Nbs[i], density=True)
    axes[i].bar(bins[:-1], hist, width=np.diff(bins), alpha=0.5, color='k', edgecolor='w')

    axes[i].set_title(f'$N_b$={Nbs[i]}, h={round(h,2)}')

plt.tight_layout()
plt.show()
```

### Partie 2 - Estimation par noyaux (en 1D)

$(x_1, ..., x_n)$ sont toujours $n$ réalisations indépendantes d'une variable aléatoire réelle.

L'une des motivations principales est de faire directement dépendre l'estimation des données, et de ne plus dépendre de découpages arbitraires de $\mathbb{R}$. Pour cela, on propose la méthodologie suivante :
- Choisir une fonction $K$ appelée "noyau". On choisit $K$ positive ou nulle, symétrique, et d'intégrale 1.
- Centrer $K$ sur chaque observation $x_i$.
- L'estimateur à noyaux est alors
$$\hat{f}_n(x) = \frac{1}{n} \sum_{i=1}^n K(x - x_i).$$

On pourra par exemple choisir des noyaux $K$ continus, dérivables partout... Propriétés dont $\hat{f}_n$ héritera.

**Q2**. En réalité, on souhaite faire dépendre cet estimateur d'un paramètre de lissage $h > 0$, qu'on appelle largeur de fenêtre (ou *bandwidth* en anglais).

Le (véritable) estimateur à noyaux, aussi appelé méthode de Parzen-Rosenblatt (du nom des deux statisticiens l'ayant développée) est :
$$\hat{f}_n(x) = \frac{1}{nh} \sum_{i=1}^n K(\frac{x - x_i}{h}).$$

Vérifier que c'est bien l'estimateur d'une densité.

Implémenter cet estimateur sur les données de la partie 1, et montrer l'influence de la valeur de $h$. On pourra encore utiliser $\texttt{plt.subplots}$. Commenter.

```python
# Implement the (true) kernel estimator and show the influence of h
h_values = [0.005, 0.1, 0.5, 1, 2]

# Créer une figure avec des subplots
fig, axes = plt.subplots(len(h_values), 1, figsize=(5, 3 * len(h_values)))

# Boucle sur les différentes valeurs de h
x = np.linspace(min(points), max(points), 1000)
for i, h in enumerate(h_values):
    # Initialiser la somme à zéro pour chaque h
    sum_result = [0 for a in x]
    
    # Calculer la somme
    for x_i in points:
        sigma = h
        y = (1/(np.sqrt(2 * np.pi))) * np.exp(-((x - x_i)/h)**2 / (2)) / N / h
        sum_result += y
    
    # Tracer le subplot
    axes[i].plot(x, sum_result, 'b--', alpha=0.5, label=f'$f_n$')
    axes[i].plot(x, pdf3, 'r', label='$f$')
    axes[i].legend()
    axes[i].set_title(f'h={h}')

# Afficher les subplots
plt.show()
```
